# Covara One: ML Exploratory Data Analysis & XGBoost Prototype

This notebook analyzes the seed data distribution, evaluates feature engineering for fraud detection, and trains the prototype models for severity prediction.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

ROOT = os.path.dirname(os.path.dirname(os.path.dirname(os.path.abspath('__file__'))))
df = pd.read_csv(os.path.join(ROOT, 'data', 'samples', 'joined_training_data_seed.csv'))
df.head()
```

## Feature Correlation

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()
```
